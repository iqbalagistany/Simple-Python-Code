```python
from random import randint

bool = True
sum_user = 0
sum_computer = 0
counter = 1

print('''SELAMAT DATANG di GAME TEBAK ANGKA
Game akan menghitung jumlah dari setiap angka yang anda masukkan
Jika jumlah angka yang anda masukkan sama dengan jumlah angka dari komupter, Maka anda akan menang!
''')

while (bool):
  print("\nGame ke-", counter)
  user = int(input("Tebakan Anda (-6 sampai 6):\t "))
  if (user < -6 or user>6):
    print("Out Of Range")
    continue
  computer = randint(-6,6)
  print("Tebakan Komputer: ", computer)
  sum_user += user
  sum_computer += computer
  print("\nJumlah Tebakan User:\t\t", sum_user, "\nJumlah Tebakan Komputer:\t", sum_computer)
  print("\n")
  counter += 1
  if sum_user == sum_computer:
    bool = False
    break
  

print("\nSelamat Anda menang pada Game ke-", counter-1)
```
